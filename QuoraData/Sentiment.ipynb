```python
import json
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis', model='roberta-large')
```

```python
with open('iiith.json') as f:
    data = json.load(f)
questions_answers = [item['selection2'] for item in data['selection1']]
```

```python
print(questions_answers)
```

```python
for question in questions_answers:
    # Perform sentiment analysis on the text
    results = classifier(question)

    # Print the sentiment and confidence score
    print(f"Sentiment: {results[0]['label']}, Score: {results[0]['score']}")
```
